```python
import telebot

API_TOKEN = '8264287493:AAHHUwi1Kysciye00NTo82Mb_KNu97Une-w'  # 🔁 Replace with your token
bot = telebot.TeleBot(API_TOKEN)

# ============================================================
# 🏥 MEDICAL DICTIONARY — Medicine ➜ Disease(s)
# ============================================================
medicine_to_disease = {
    # Antibiotics
    "Amoxicillin":       ["Bacterial infections", "Pneumonia", "Ear infections", "Tonsillitis", "UTI"],
    "Azithromycin":      ["Bacterial infections", "Pneumonia", "Sinusitis", "Bronchitis", "STIs"],
    "Ciprofloxacin":     ["UTI", "Typhoid fever", "Bacterial diarrhea", "Skin infections"],
    "Metronidazole":     ["Bacterial vaginosis", "Trichomoniasis", "Amoebic dysentery", "Giardiasis"],
    "Doxycycline":       ["Malaria (prevention)", "Lyme disease", "Acne", "Chlamydia", "Typhus"],
    "Penicillin":        ["Strep throat", "Syphilis", "Dental infections", "Scarlet fever"],
    "Clindamycin":       ["Skin infections", "Dental infections", "Bone infections", "Pneumonia"],
    "Erythromycin":      ["Respiratory infections", "Acne", "Pertussis", "Legionnaires disease"],
    "Cephalexin":        ["Skin infections", "UTI", "Respiratory tract infections", "Bone infections"],
    "Tetracycline":      ["Acne", "Chlamydia", "Lyme disease", "H. pylori infection"],

    # Antivirals
    "Acyclovir":         ["Herpes simplex", "Chickenpox", "Shingles", "Herpes zoster"],
    "Oseltamivir":       ["Influenza (Flu)", "H1N1 swine flu"],
    "Remdesivir":        ["COVID-19", "Severe viral infections"],
    "Ribavirin":         ["Hepatitis C", "RSV", "Hemorrhagic fevers"],
    "Valacyclovir":      ["Herpes simplex", "Shingles", "Cold sores", "Genital herpes"],

    # Antimalarials
    "Chloroquine":       ["Malaria", "Lupus", "Rheumatoid arthritis"],
    "Artemether":        ["Malaria", "Plasmodium falciparum"],
    "Quinine":           ["Malaria", "Leg cramps"],
    "Hydroxychloroquine":["Malaria", "Lupus", "Rheumatoid arthritis"],

    # Pain / Anti-inflammatory
    "Paracetamol":       ["Fever", "Headache", "Body pain", "Common cold", "Toothache"],
    "Ibuprofen":         ["Pain", "Fever", "Inflammation", "Arthritis", "Menstrual cramps"],
    "Aspirin":           ["Pain", "Fever", "Heart attack prevention", "Blood clots", "Inflammation"],
    "Naproxen":          ["Arthritis", "Back pain", "Menstrual pain", "Gout", "Tendinitis"],
    "Diclofenac":        ["Arthritis", "Muscle pain", "Migraine", "Ankylosing spondylitis"],
    "Celecoxib":         ["Osteoarthritis", "Rheumatoid arthritis", "Menstrual pain", "Colon polyps"],
    "Tramadol":          ["Moderate to severe pain", "Fibromyalgia"],
    "Morphine":          ["Severe pain", "Cancer pain", "Heart attack pain"],
    "Codeine":           ["Mild to moderate pain", "Cough", "Diarrhea"],

    # Cardiovascular
    "Atorvastatin":      ["High cholesterol", "Cardiovascular disease prevention"],
    "Amlodipine":        ["Hypertension", "Angina", "Coronary artery disease"],
    "Lisinopril":        ["Hypertension", "Heart failure", "Kidney protection in diabetes"],
    "Metoprolol":        ["Hypertension", "Angina", "Heart failure", "Arrhythmia"],
    "Warfarin":          ["Blood clots", "Deep vein thrombosis", "Pulmonary embolism", "Atrial fibrillation"],
    "Clopidogrel":       ["Heart attack prevention", "Stroke prevention", "Atherosclerosis"],
    "Digoxin":           ["Heart failure", "Atrial fibrillation"],
    "Furosemide":        ["Heart failure", "Edema", "Hypertension", "Kidney disease"],
    "Spironolactone":    ["Heart failure", "Hypertension", "Edema", "Hyperaldosteronism"],
    "Losartan":          ["Hypertension", "Heart failure", "Diabetic nephropathy"],
    "Nitroglycerin":     ["Angina", "Heart failure", "Chest pain"],

    # Diabetes
    "Metformin":         ["Type 2 diabetes", "PCOS", "Prediabetes"],
    "Insulin":           ["Type 1 diabetes", "Type 2 diabetes", "Diabetic ketoacidosis"],
    "Glibenclamide":     ["Type 2 diabetes"],
    "Sitagliptin":       ["Type 2 diabetes"],
    "Empagliflozin":     ["Type 2 diabetes", "Heart failure", "Chronic kidney disease"],
    "Liraglutide":       ["Type 2 diabetes", "Obesity"],

    # Respiratory
    "Salbutamol":        ["Asthma", "COPD", "Bronchospasm"],
    "Budesonide":        ["Asthma", "COPD", "Nasal polyps", "Crohn's disease"],
    "Montelukast":       ["Asthma", "Allergic rhinitis", "Exercise-induced asthma"],
    "Ipratropium":       ["COPD", "Asthma", "Chronic bronchitis"],
    "Fluticasone":       ["Asthma", "Allergic rhinitis", "COPD"],
    "Theophylline":      ["Asthma", "COPD", "Neonatal apnea"],

    # Gastrointestinal
    "Omeprazole":        ["GERD", "Peptic ulcer", "H. pylori", "Zollinger-Ellison syndrome"],
    "Ranitidine":        ["GERD", "Peptic ulcer", "Gastritis"],
    "Ondansetron":       ["Nausea", "Vomiting", "Chemotherapy-induced nausea"],
    "Loperamide":        ["Diarrhea", "IBS"],
    "Pantoprazole":      ["GERD", "Peptic ulcer", "Esophagitis"],
    "Lactulose":         ["Constipation", "Hepatic encephalopathy"],
    "Domperidone":       ["Nausea", "Gastroparesis", "GERD"],
    "Metoclopramide":    ["Nausea", "Gastroparesis", "Migraine-related vomiting"],
    "Mesalamine":        ["Ulcerative colitis", "Crohn's disease"],

    # Mental Health / Neurology
    "Sertraline":        ["Depression", "OCD", "PTSD", "Panic disorder", "Social anxiety"],
    "Fluoxetine":        ["Depression", "OCD", "Bulimia nervosa", "Panic disorder"],
    "Amitriptyline":     ["Depression", "Neuropathic pain", "Migraine prevention", "Insomnia"],
    "Diazepam":          ["Anxiety", "Seizures", "Muscle spasms", "Alcohol withdrawal"],
    "Alprazolam":        ["Anxiety", "Panic disorder"],
    "Risperidone":       ["Schizophrenia", "Bipolar disorder", "Autism-related irritability"],
    "Lithium":           ["Bipolar disorder", "Mania", "Depression"],
    "Haloperidol":       ["Schizophrenia", "Bipolar disorder", "Tics (Tourette)"],
    "Levodopa":          ["Parkinson's disease"],
    "Donepezil":         ["Alzheimer's disease", "Dementia"],
    "Methylphenidate":   ["ADHD", "Narcolepsy"],
    "Valproate":         ["Epilepsy", "Bipolar disorder", "Migraine prevention"],
    "Carbamazepine":     ["Epilepsy", "Trigeminal neuralgia", "Bipolar disorder"],
    "Phenytoin":         ["Epilepsy", "Seizures"],
    "Gabapentin":        ["Neuropathic pain", "Epilepsy", "Restless legs syndrome"],
    "Pregabalin":        ["Neuropathic pain", "Fibromyalgia", "Epilepsy", "Anxiety"],

    # Allergy / Immune
    "Cetirizine":        ["Allergic rhinitis", "Hives", "Hay fever", "Skin allergies"],
    "Loratadine":        ["Allergic rhinitis", "Hives", "Chronic urticaria"],
    "Fexofenadine":      ["Allergic rhinitis", "Chronic urticaria"],
    "Prednisolone":      ["Asthma", "Allergies", "Inflammatory diseases", "Autoimmune disorders"],
    "Dexamethasone":     ["Inflammation", "Allergic reactions", "COVID-19 (severe)", "Cancer treatment"],
    "Hydrocortisone":    ["Eczema", "Allergic reactions", "Adrenal insufficiency"],
    "Epinephrine":       ["Anaphylaxis", "Severe allergic reaction", "Cardiac arrest"],

    # Hormonal
    "Levothyroxine":     ["Hypothyroidism", "Thyroid cancer (post-surgery)"],
    "Propylthiouracil":  ["Hyperthyroidism", "Graves' disease"],
    "Tamoxifen":         ["Breast cancer", "Gynecomastia"],
    "Testosterone":      ["Hypogonadism", "Delayed puberty", "Gender dysphoria"],
    "Estrogen":          ["Menopause symptoms", "Osteoporosis", "Hypogonadism"],

    # Dermatology
    "Tretinoin":         ["Acne", "Wrinkles", "Skin discolouration"],
    "Benzoyl Peroxide":  ["Acne"],
    "Ketoconazole":      ["Fungal infections", "Dandruff", "Candidiasis"],
    "Terbinafine":       ["Fungal infections", "Athlete's foot", "Ringworm", "Nail fungus"],
    "Clotrimazole":      ["Fungal infections", "Vaginal yeast infection", "Tinea"],
    "Permethrin":        ["Scabies", "Head lice"],

    # Vitamins / Supplements
    "Vitamin D":         ["Vitamin D deficiency", "Osteoporosis", "Rickets", "Immune support"],
    "Vitamin C":         ["Scurvy", "Immune support", "Iron absorption"],
    "Folic Acid":        ["Folate deficiency", "Neural tube defect prevention", "Anemia"],
    "Iron":              ["Iron deficiency anemia"],
    "Vitamin B12":       ["B12 deficiency", "Pernicious anemia", "Neuropathy"],
    "Calcium":           ["Osteoporosis", "Calcium deficiency", "Hypocalcemia"],
    "Zinc":              ["Zinc deficiency", "Immune support", "Wound healing"],

    # Oncology
    "Methotrexate":      ["Rheumatoid arthritis", "Psoriasis", "Leukemia", "Breast cancer"],
    "Cyclophosphamide":  ["Cancer", "Lupus", "Lymphoma"],
    "Tamoxifen":         ["Breast cancer"],
    "Imatinib":          ["CML (Leukemia)", "Gastrointestinal stromal tumors"],

    # Antiparasitic
    "Albendazole":       ["Worm infections", "Hydatid disease", "Neurocysticercosis"],
    "Ivermectin":        ["River blindness", "Scabies", "Strongyloidiasis", "Head lice"],
    "Praziquantel":      ["Schistosomiasis", "Tapeworm infections"],
    "Mebendazole":       ["Pinworm", "Roundworm", "Hookworm", "Whipworm"],

    # Anticoagulants
    "Heparin":           ["Deep vein thrombosis", "Pulmonary embolism", "Atrial fibrillation"],
    "Rivaroxaban":       ["DVT", "Pulmonary embolism", "Stroke prevention"],
    "Apixaban":          ["Atrial fibrillation", "DVT", "Pulmonary embolism"],

    # Ophthalmology
    "Latanoprost":       ["Glaucoma", "Ocular hypertension"],
    "Timolol":           ["Glaucoma", "Ocular hypertension"],
    "Chloramphenicol":   ["Bacterial conjunctivitis (eye drops)", "Eye infections"],

    # Urology
    "Tamsulosin":        ["BPH (Enlarged prostate)", "Kidney stones passage"],
    "Sildenafil":        ["Erectile dysfunction", "Pulmonary arterial hypertension"],
    "Finasteride":       ["BPH (Enlarged prostate)", "Male pattern baldness"],

    # Misc
    "ORS":               ["Dehydration", "Diarrhea", "Cholera"],
    "Activated Charcoal":["Poisoning", "Drug overdose"],
    "Naloxone":          ["Opioid overdose"],
    "Flumazenil":        ["Benzodiazepine overdose"],
}

# ============================================================
# 🔁 AUTO-BUILD Reverse Dictionary — Disease ➜ Medicine(s)
# ============================================================
disease_to_medicine = {}
for medicine, diseases in medicine_to_disease.items():
    for disease in diseases:
        disease_lower = disease.lower()
        if disease_lower not in disease_to_medicine:
            disease_to_medicine[disease_lower] = []
        disease_to_medicine[disease_lower].append(medicine)


# ============================================================
# 🔍 SMART FUZZY SEARCH
# ============================================================
def fuzzy_find(query, dictionary):
    """Return exact or partial matches from a dictionary."""
    query_lower = query.lower().strip()
    # Exact match
    if query_lower in dictionary:
        return query_lower, dictionary[query_lower]
    # Partial match
    matches = {k: v for k, v in dictionary.items() if query_lower in k or k in query_lower}
    if matches:
        best = max(matches, key=lambda k: len(k))
        return best, matches[best]
    return None, None


def search_medical(query):
    """Search medicine → disease OR disease → medicine."""
    # Check medicine dictionary (case-insensitive)
    med_lower = query.lower().strip()
    med_match = next((k for k in medicine_to_disease if k.lower() == med_lower), None)

    if not med_match:
        # partial medicine match
        med_match = next((k for k in medicine_to_disease if med_lower in k.lower()), None)

    if med_match:
        diseases = medicine_to_disease[med_match]
        return "medicine", med_match, diseases

    # Check disease dictionary
    disease_key, medicines = fuzzy_find(query, disease_to_medicine)
    if disease_key:
        return "disease", disease_key.title(), medicines

    return None, None, None


# ============================================================
# 🤖 BOT HANDLERS
# ============================================================
@bot.message_handler(commands=['start'])
def send_welcome(message):
    bot.reply_to(message, (
        "👨‍⚕️ *Welcome to MediBot!*\n\n"
        "I can help you with:\n"
        "💊 *Medicine → Disease*: Type a medicine name\n"
        "🏥 *Disease → Medicine*: Type a disease name\n\n"
        "You can also search multiple at once — separate with commas!\n\n"
        "_Example:_ `Paracetamol, Diabetes, Aspirin`\n\n"
        "Type /help for more info."
    ), parse_mode="Markdown")


@bot.message_handler(commands=['help'])
def send_help(message):
    bot.reply_to(message, (
        "📖 *How to use MediBot:*\n\n"
        "• Type any *medicine name* → get diseases it treats\n"
        "• Type any *disease name* → get medicines for it\n"
        "• Separate multiple queries with *commas*\n\n"
        "*Examples:*\n"
        "`Amoxicillin` → shows diseases\n"
        "`Diabetes` → shows medicines\n"
        "`Fever, Metformin` → shows both!\n\n"
        "⚠️ *Disclaimer:* This bot is for educational purposes only. "
        "Always consult a licensed doctor before taking any medication."
    ), parse_mode="Markdown")


@bot.message_handler(commands=['list'])
def send_list(message):
    meds = ", ".join(sorted(medicine_to_disease.keys()))
    bot.reply_to(message, f"💊 *Available Medicines ({len(medicine_to_disease)}):*\n\n{meds}", parse_mode="Markdown")


@bot.message_handler(func=lambda message: True)
def handle_query(message):
    queries = [q.strip() for q in message.text.split(',') if q.strip()]
    responses = []

    for query in queries:
        if not query:
            continue

        result_type, match, results = search_medical(query)

        if result_type == "medicine":
            disease_list = "\n".join([f"   • {d}" for d in results])
            responses.append(
                f"💊 *Medicine:* {match}\n"
                f"🏥 *Treats:*\n{disease_list}"
            )

        elif result_type == "disease":
            medicine_list = "\n".join([f"   • {m}" for m in results])
            responses.append(
                f"🏥 *Disease/Condition:* {match}\n"
                f"💊 *Common Medicines:*\n{medicine_list}"
            )

        else:
            responses.append(
                f"❓ *'{query}'* not found in database.\n"
                f"   Try a different spelling, or type /list to see all medicines."
            )

    divider = "\n\n" + "─" * 30 + "\n\n"
    full_response = divider.join(responses)
    full_response += (
        "\n\n⚠️ _For educational use only. Consult a doctor before use._"
    )

    bot.reply_to(message, full_response, parse_mode="Markdown")


# ============================================================
# 🚀 START BOT
# ============================================================
print("🤖 MediBot is running...")
bot.polling(none_stop=True)
```
